```python
import requests
from bs4 import BeautifulSoup

url = 'https://movies.yahoo.com.tw/name_movies/1217?type=2'
url2 = 'https://movies.yahoo.com.tw/movieinfo_main/%E7%8E%A9%E5%91%BD%E9%97%9C%E9%A0%AD9-fast-furious-9-10482'
r = requests.get(url)
soup = BeautifulSoup(r.text)

infos = []
imgs = []
soup
```

```python
visit_list = []
for b in soup.find_all('div', class_="release_movie_name"):
    visit_list.append(b.find('a')['href'])
visit_list
```

```python
info = {}
infos = []
res = []
for i in visit_list:
    r = requests.get(i)
    soup = BeautifulSoup(r.text)
    info = {}
    info['名稱'] = soup.find('div', class_="movie_intro_info_r").find('h1').text.strip()
    tag = []
    for j in soup.find('div', class_ = 'level_name_box').find_all('div', class_ = 'level_name'):
        tag.append(j.find('a', class_ = "gabtn").text.strip())
    info['分類'] = tag
    info['劇情介紹'] =  soup.find('div', class_ = 'gray_infobox_inner').find('span').text.strip()
    date = soup.find('div', class_ = 'movie_intro_info_r').find('span').text.strip()
    date = date.split('：')
    info['上映日期'] = date[1]
    print(info)
    infos.append(dict(list(info.items())))
```

```python
import pandas as pd

df = pd.DataFrame(infos)
df.to_csv('movies.csv', encoding='utf_8_sig')

df
```
